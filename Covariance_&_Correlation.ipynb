## 공분산

* 두 개 이상의 확률변수가 어떻게 퍼져있는지를 나타내는 것
* 한 개의 확률변수 X에 대한 분포를 파악할 때, 평균과 분산을 이용
* 두 개의 확률변수 X와 Y에 대한 분포를 파악할 때, X와 Y의 평균 및 분산을 이용
  * 두 가지 확률변수가 어떻게 퍼져있는지를 파악하여 둘 이상의 변수 간의 상관관계를 파악할 수 있다.
  * 함께 같이 퍼져있는 경우 강한 상관관계를 갖게 된다. 




image.png

* 확률변수 X와 Y의 평균을 $\mu, v$라 할 때, $E((X-\mu)(Y-v))$로 표현할 수 있음
* 한마디로 편차 간의 곱의 평균으로 표현할 수 있음
* $cov(X, Y) = E(XY) - \mu v$
* X와 Y 변수가 독립일 경우 공분산은 0이 된다. -> 겹치는 부분이 없고 분산에도 어떠한 규칙이나 패턴이 없다는 것
* 상관계수 수식 -> $p = \frac {cov(X,Y)}{\sqrt var(X) var(Y)}$ 공분산에서 각 변수의 분산의 곱의 제곱근을 나눈 값을 통해 상관계수 파악가능
